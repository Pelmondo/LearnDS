```python
index_word_sentens = {}
```

```python
k = 0
for i in clear_sents:
    se = set(i)
    for j in range(len(i)):
        if i[j] in se:
            if i[j] in index_word_sentens[k].kyes():
                index_word_sentens[k][i[j]] += 1
            else:
                index_word_sentens[k][i[j]] = 1
    k += 1
index_word_sentens    
```
