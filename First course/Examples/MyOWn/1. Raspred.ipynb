```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)

norm_rv.rvs(size=10)
```

```python
x = np.linspace(0,4,100)
cdf = (norm_rv.cdf(x))
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
a = 1
b = 4

uniform_rv = sts.uniform(a, b-a)
uniform_rv.rvs(10)
```

```python
x = np.linspace(0,5,100)
cdf = uniform_rv.cdf(x)
plt.plot(x, cdf)
```

```python
x = np.linspace(0,5,100)
pdf = uniform_rv.pdf(x)
plt.plot(x, pdf);
```

```python
bernoulli_rv = sts.bernoulli(0.7)
bernoulli_rv.rvs(10)
```

```python
binomial_rv = sts.binom(20, 0.7)
binomial_rv.rvs(10)
```

```python
x = np.linspace(0,20,21)
cdf = binomial_rv.cdf(x)
plt.step(x, cdf)
```

```python
pmf = binomial_rv.pmf(x)
plt.plot(x, pmf, 'x')
```

```python
x = np.linspace(0,45,46)
for N in [20,30]:
    for p in [0.2, 0.7]:
        rv = sts.binom(N, p)
        cdf = rv.cdf(x)
        plt.step(x, cdf, label='$N=%s, p=%s$' % (N,p))
plt.legend()
```
